From Griffith's Sec. 9.4, especially the discussion around Eq. (9.167), the intensity of an EM wave with wavelength $\lambda$ decreases with distance due to absorption according to,


\begin{align}
    I(z, \lambda) \propto e^{- \alpha(\lambda) z}
\end{align}


where $\alpha(\lambda)$ is the absorption coefficient.  As we know in QM, the intensity of a classical wave is equivalent to the probability of a QM wave.  Therefore knowing $I(z,\lambda) \propto \exp[-\alpha(\lambda) z]$ implies the same thing about the number of photons which survive a distance $z$:

\begin{align}
    N(z, \lambda) \propto e^{- \alpha(\lambda) z}   \: .
\end{align}

This is the solution to the differential equation,

\begin{align}
    \frac{dN}{dz} = - \alpha N(z)
\end{align}

which tells us to interpret $\alpha$ as the $\textbf{absorption probability per unit length}$! 

Therefore if at a position $z$ in the proton range calculation we have $N(z)$ photons, then,  

\begin{align}
    \ \mathrm{prior \: photons \: lost} = \frac{dN}{N} = - \alpha \: dz
\end{align}

This tells us that:

At every step $dx$ in the proton range integration, we should $\textbf{remove a percentage of photons}$ equal to $\alpha dx$. This will continuously reduce the intensity of emitted Cerenkov photons. Interestingly, there is a nontrivial dependence on the position $x$.  There are more photons radiated $\textit{early}$, but those also suffer more $\textit{attenuation}$. But photons radiated $\textit{later}$ are $\textit{more likely to survive}$. This also depends on the $\textit{wavelength}$ of the photon, which further determines the Cerenkov angle of that emission. Seems like a $\textit{highly nontrivial interplay}$ between the $\textbf{velocity, distance, wavelength, and angle}$.  Cool!




```python
#Refractive index new fit: M. Sievert (Mathematica equivalent)
# initially
# a_0 = 1.189
# a_pvs = 0.00029
# tau = 0.0005

# {a0 -> 1.18416, aPVS -> 0.000325985, \[CapitalGamma] -> 0.000693652}
a_0 = 1.18416
a_pvs = 0.000325985
gamma = 0.000693652


lamd = np.linspace(0,500,2000)   #wavelength of light 
lamd_uv = 106.6 #lambda_UV = 106.6 nm
lamd_ir = 908.3 #lambda_IR = 908.3 nm 
lamd_peak = 115.11 #where n is max 


#Formula: 
n_pvs = a_0 + (a_pvs * ((lamd_uv**(-1) - lamd**(-1))/((lamd_uv**(-1) - lamd**(-1))**2 + gamma**2)))

#absorption co-efficient, alpha 

sigma = (lamd_peak - lamd_uv)
mu = lamd_peak
abs_coeff = np.exp(-(lamd - lamd_peak)**2/(2*sigma**2))/(np.sqrt(2*np.pi*sigma**2))

# plt.plot(n_pvs)

#Plot of n vs lambda in the LAr medium.

fig = plt.figure(figsize=(12,12))

ax = fig.add_subplot(111)
ax.set_title("Refractive index vs wavelength for LAr: Our fit", fontsize=22)
ax.plot(lamd, n_pvs, linewidth = 3.5, label="$n$ vs $\lambda$  for our fit")
# ax.plot(lamd, y, linewidth = 3.5,label="$1/ \\beta$")

ax.tick_params(axis='both',which='major',labelsize=20,length=10)

ax.set_xlabel("wavelength, $\\lambda$ (nm)",fontsize=24)
ax.set_ylabel('$refractive index$ , $n$',fontsize=24)
ax.grid()
ax.legend(fontsize=18)

n_pvs,sigma,abs_coeff
```

Proposal:
*Trying to incorporate the lambda dependence of absorption will be complicated
*We already have an upper bound on the number of photons (by having no absorption)
*Let’s estimate a lower bound on the number of photons (by assuming the absorption is constant)
*So try taking alpha = the peak of the absorption curve (about 0.1 $nm^{-1}$)
*Then we can attenuate the photons as described on Slides 2-3 using the constant alpha
*If we wanted to, we could also use an “average” value of (alpha in some sense)
*Going beyond this will require folding in 𝛼(𝜆) into a step-by-step evolution in dx of 𝑑𝑁/(𝑑𝑥 𝑑𝜆).
